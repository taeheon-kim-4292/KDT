#### [ Data Frame 연결/통합]
- 통합
    * DF에서 특정 컬럼을 기준으로 통합해주는 기능
    * 함수 : pandas.merge()
    

```python
# [ 1 ] 모듈로딩
import pandas as pd
```

```python
# [2] 데이터 준비
file_name='stock price.xlsx'
file_name2='stock valuation.xlsx'
```

```python
# [3] XLSX 파일 ===> DF로 저장
priceDF=pd.read_excel(file_name)
valDF=pd.read_excel(file_name2)
```

```python
print(priceDF.head(3), valDF.head(3), sep='\n\n')
```

```python
# [4] DF 합치기 -> pandas.merge()함수
stockDF=pd.merge(priceDF,valDF)
stockDF.info()
```

```python
# - DF별 특정 컬럼 지정
stockDF=pd.merge(priceDF, valDF,left_on='stock_name',right_on='name')
```

```python
stockDF
```

```python
# [5] DF 합치기 ==> DataFrame.join() 메서드
priceDF.join(valDF,lsuffix='_x',rsuffix='_y')
```

```python
priceDF.join(priceDF,lsuffix='_x',rsuffix='_y')
```
