### DataFrame에서 행/열 추가
    - 행 방향 추가
    - 열 방향 추가

```python
# 모듈 로딩
import pandas as pd
```

```python
# 데이터 준비
data={'name':['홍길동','마징가','원더우먼'],
      'job' :['의적','로봇','영웅'],
      'gender':['남','남','여'],
      'age': [100,200,300],
      'height':[180,300,200]}
```

```python
# DF로 데이터 저장
dataDF=pd.DataFrame(data)
dataDF
```

- [실습]  행 추가: 변수명.loc['추가할 행이름'] = 데이터**

    - .iloc[정수] --> ERROR !!

    - if 데이터수 == 컬럼수 : 행추가 가능 

    - if 데이터수 == 1개 : 행추가 가능

    - 그 외 모두 ERROR

```python
## 1개 행 추가 => 데이터 수가 열 갯수와 동일한 경우
dataDF.loc[3]=['베트맨','박쥐','남',40,190]
```

```python
# 데이터 수를 1개로 추가하면 모든 컬럼에 동일한 값 저장됨
dataDF.loc[4]=['베트맨','박쥐','남',40,190]
dataDF.loc[5]=100
dataDF
```

```python
## 1개 행 추가 => 데이터 수를 2개로 = ERROR
# dataDF.loc[6]=['스파이더맨','거미']
# dataDF            => ERROR
```

- [실습]  열 추가: 변수명['추가할 열이름'] = 데이터


    - if 데이터수 == 행 수 : 열추가 가능 

    - if 데이터수 == 1개 : 열추가 가능

    - 그 외 모두 ERROR

- [실습] 열/컬럼 삭제 : 변수명.drop(   ,axis=1 또는 axis='columns') 메서드 *axis=0은 행이고 1이면 열
    
    - 삭제 방향 즉, 기준을 열로 설정 필수

    -변수명.drop(labels, axis=1)  : axis의 설정 값에 따라 행 또는 열 삭제 진행
    
    -변수명.drop(columns) : 지정된 인덱스의 열 삭제

```python
## 컬럼/열 추가 ==> weight 추가 : 행 개수 == 데이터 개수
dataDF['weight']=[90,100,80,78,78,90]
dataDF
```

```python
## 컬럼/열 추가 ==> weight 추가 : 1개 == 데이터 개수
dataDF['weight2']=0
dataDF
```
