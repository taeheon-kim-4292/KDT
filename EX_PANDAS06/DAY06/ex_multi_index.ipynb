### [멀티인덱스]
 - 2개 이상의 행과 열로 구성된 인덱스
 

```python
# [1] 모듈 로딩
import pandas as pd
import df_util as utill
```

```python
# [2] 데이터 준비
file1 = 'test_data.csv'
file2 = 'test_data2.csv'
```

```python
# [3] csv ==> DF로 저장
DF1 = pd.read_csv(file1)
DF2 = pd.read_csv(file2)
```

```python
# [4] 데이터 확인 => 기본 정보
utill.checkDataFrame(DF1, 'DF1')
utill.checkDataFrame(DF2, 'DF2')
```

```python
# [5] 행 인덱스 설정
# [5-1] DF1의 첫번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])
df11=DF1.set_index('index')
df11.index.name=None
df11.index

# 'A' 행 데이터 추출
df11.loc['A']
```

```python
# 'A' , 'C' 행 추출
df11.loc[['A','C']]
```

```python
# [5] 행 인덱스 설정
# [5-2] DF1의 첫번째, 두번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])
df12=DF1.set_index(['index','col1'])
df12.index

```

```python
# [5] 행 인덱스 설정
# [5-2] DF1의 첫번째, 두번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])
df12=DF1.set_index(['index','col1'])
df12.index

# ('A',8) ,('A',1)행 추출
df12.loc[[('A',8),('A',1)]]
```

```python
# 멀티인덱스의 속성들 확인
mIdx=df12.index 

print(f'nlevels : {mIdx.nlevels}')
print(f'levels : {mIdx.levels}')
print(f'names : {mIdx.names}')

# mIdx.names=[None,None]
# print(f'names : {mIdx.names}')
# mIdx.names[0]='A'
mIdx.set_names('a',level=0)
```

```python
# 멀티인덱스에서 level0이 동일한 행 추출
# - 'A' 
df12.loc['A']
```

```python
# 멀티인덱스에서 level1이 동일한 행 추출
# - 8
# df12.loc[8]       -->Error 발생
df12.xs(8,level=1)
```

```python
# 멀티인덱스에서 행추가
# => 인덱스에 행 추가 ==> 변수명.loc[행 인덱스] = 값
df12.loc[('D',9), :] = [9,9]
df12
```
