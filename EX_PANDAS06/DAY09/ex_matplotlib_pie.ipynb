#### [ 파이 그래프 ]
    - 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA =r'C:\Users\kdp\Desktop\KDT\EX_PANDAS06\data\auto_mpg.csv'
```

```python
# [3] CSV -> DF 저장
dataDF=pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터의 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
originSR=dataDF['origin'].value_counts()
x=originSR.index.to_list()      # 시리즈라서 to_list
h=originSR.values_.tolist()
```

```python
dataDF['origin'].unique()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
originBar=plt.bar(x,h,color=['red','blue','yellow'],edgecolor=['blue'])

# 막대에 라벨을 설정
plt.bar_label(originBar,h, label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3],['USA','Europe','Japan'], rotation=45, fontsize=15)
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - 파이 그래프
plt.pie(h,labels=['USA','Europe','Japan'], autopct='%.1f%%')
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - 파이 그래프
plt.pie(h,labels=['USA','Europe','Japan'], autopct='%.1f%%',counterclock=False, startangle=90)
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - 파이 그래프
plt.pie(h,labels=['USA','Europe','Japan'], autopct='%.1f%%',counterclock=False, startangle=90
        ,explode=[0,5,0])
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - 파이 그래프
plt.pie(h,labels=['USA','Europe','Japan'], autopct='%.1f%%',counterclock=False, startangle=90
        ,wedgeprops={'linewidth':6, 'width':0.9, 'edgecolor':'w'})
plt.show()
```
