```python
# 모듈 준비
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
# 한글 폰트 설정 ->폰트 매니저 모듈
from matplotlib import font_manager as fm
from matplotlib import rc
import seaborn as sns
```

```python
# 데이터 불러오기
data=pd.read_csv(r'C:\Users\kdp\Desktop\KDT\EX_PANDAS06\data\blood7.csv', encoding='euc-kr')
data
```

```python
# 빈혈코드 삭제
data['사망원인별(104항목)'] = data['사망원인별(104항목)'].str.replace(r' \(D50-D64\)', '',regex=True)
data
```

```python
# 중복 컬럼 삭제
data = data.drop(0)
data
```

```python
# 컬럼명 재설정
data=data.rename(columns={'사망원인별(104항목)':'사망원인', '2022':'2022_사망자수', '2022.1':'2022_사망률', '2021':'2021_사망자수', '2021.1':'2021_사망률',
       '2020':'2020_사망자수', '2020.1':'2020_사망률', '2019':'2019_사망자수', '2019.1':'2019_사망률', '2018':'2018_사망자수',
         '2018.1':'2018_사망률', '2017':'2017_사망자수', '2017.1':'2017_사망률'})
```

```python
# 결측치 확인
data.info()
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
df=data.drop(data.columns[0],axis=1)
df
```

```python
df[df.columns[3:]]=df[df.columns[3:]].astype(float)
df.info()
```

```python
# 년도별 컬럼을 생성
years = ['2022', '2021', '2020', '2019', '2018', '2017']

# 각 년도 사망자 수를 계산
for year in years:
    df[f'{year}_사망자수'] = df[f'{year}_사망자수']

# 시도별로 그룹화하고 각 년도의 사망자 수를 합산
result = df.groupby('시도별')[['2022_사망자수', '2021_사망자수', '2020_사망자수', 
                              '2019_사망자수', '2018_사망자수', '2017_사망자수']].sum()
result
```

```python
result=result.rename(columns={'2022_사망자수':'2022','2021_사망자수':'2021','2020_사망자수':'2020','2019_사망자수':'2019','2018_사망자수':'2018','2017_사망자수':'2017'})
result.index.name='년'
result.T
```

```python
# 데이터 불러오기
hotdata=pd.read_csv(r'C:\Users\kdp\Desktop\KDT\EX_PANDAS06\data\hot6.csv', encoding='euc-kr')
hotdata
```

```python
# '년' 열에서 불필요한 탭 문자 제거
hotdata['년'] = hotdata['년'].str.strip()

# NaN 값이 있는 행 제거
hotdata = hotdata.dropna()

# 수정된 데이터 출력
hotdata
```

```python
hotdata=hotdata.set_index('년')
hotdata
```

```python
hotdata.sort_index(ascending=False)
```

```python
dataDF=hotdata.join(result.T)
dataDF
```

```python
dataDF.describe()
```

```python
dataDF.corr()
```

```python
# 적용할 폰트 파일
font_file=r'C:\Windows\Fonts\batang.ttc'

# 폰트 패밀리
font_name= fm.FontProperties(fname=font_file).get_name()

# 새로운 폰트 패밀리 이름 지정
rc('font', family=font_name)
```

```python
fig = plt.figure(figsize=(8,8)) # Figure 생성 
fig.set_facecolor('white') # Figure 배경색 지정
ax1 = fig.add_subplot() # axes 생성
 
colors = sns.color_palette('summer', len(dataDF.index)) # 바 차트 색상
 
xtick_label_position = list(range(len(dataDF.index))) # x축 눈금 라벨이 표시될 x좌표
ax1.set_xticks(xtick_label_position) # x축 눈금 
ax1.set_xticklabels(dataDF.index) # x축 눈금 라벨
ax1.bar(xtick_label_position, dataDF['부산광역시'], color=colors) ## 바차트 출력
 
color = 'blue'
ax2 = ax1.twinx() # 새로운 axis 생성
ax2.plot(xtick_label_position, dataDF['연간폭염일수'], color=color, linestyle='--', marker='o') # 선 그래프 
ax2.tick_params(axis='y', labelcolor=color) # 눈금 라벨 색상 지정
 
plt.title('부산광역시 연간폭염일수 추이에 따른 빈혈 사망자수', fontsize=17)
plt.show()
```

- 기상청에 따르면 부산지역 2020년 여름철 평균기온은 평년(23.9±0.3℃)보다 0.5~1.5℃ 높고, 열대야 일수도 12~17일 정도로 평년과 작년에 비해 많았다.

-출처 : 글로벌뉴스통신GNA(http://www.globalnewsagency.kr)
