```python
import pandas as pd
import numpy as np
```

```python
file_name='titanic.csv'
```

```python
titanDF=pd.read_csv(file_name)
```

```python
# [4] 데이터 확인
# - 기본 정보 확인
titanDF.info()
```

```python
# - 컬럼과 실제 데이터 타입 확인
print(titanDF.head(), titanDF.tail(), sep='\n\n')
```

```python
# - 컬럼별 데이터 분포 확인
titanDF.describe()
```

```python
# [5] 데이터 전처리
# [5-1] 결측치 체크
titanDF.isna().sum()

```

```python
# 결측값 2개 제거
# (1) - 892개 데이터중 2개는 결과값에 영향을 안미친다고 판단
titanDF = titanDF.dropna(subset=['embark_town'])
titanDF
```

```python
# 형제 배우자수 + 부모 / 자녀수 + 본인자신 더해서 가족규모 파악
fam_df = titanDF['sibsp'] + titanDF['parch'] + 1

```

```python
# 대가족 기준은 4명으로 설정
Large_fam= fam_df>=4
Large_fam
```

```python
# 대가족 데이터 추출
Real_Large_fam=titanDF[Large_fam]
print(Real_Large_fam)
```

```python
# 대가족이 사는 embark_town 분포 확인
where_family = Real_Large_fam['embark_town'].value_counts()

where_family
```

```python
print("대가족이 많은 동네 분포:")
print(where_family)
```
